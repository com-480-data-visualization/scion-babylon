```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
import pandas as pd
df = pd.read_csv('international_bestsellers.csv')
out_dir = 'scion-babylon/assets'
import os
os.makedirs(out_dir, exist_ok=True)
```

```python
print(df.isnull().sum().to_string())
```

```python
df = df.dropna()
```

```python
print("### Basic Dataset Info")
print(f"Total Rows: {len(df)}")
print(f"Total Columns: {len(df.columns)}")
print(f"Columns: {list(df.columns)}\n")
```

```python
print("### Missing Values")
print(df.isnull().sum().to_string())
print("\n")
```

```python
print("### Date Range")
print(f"Start: {df['date'].min()}, End: {df['date'].max()}\n")
```

```python

print("### Unique Entities")
print(f"Total unique authors: {df['author'].nunique()}")
print(f"Total unique books: {df['title'].nunique()}")
print(f"Total unique publishers: {df['publisher'].nunique()}")
print(f"Total unique countries (list source): {df['country'].nunique()}\n")
```

```python
print("### Top 10 Authors")
print(df['author'].value_counts().head(10).to_string())
print("\n")

print("### Top 10 Books")
print(df['title'].value_counts().head(10).to_string())
print("\n")

print("### Top 10 Nationalities")
print(df['nationality'].dropna().str.split('; ').explode().value_counts().head(10).to_string())
print("\n")
```

## Plots

```python

# Set seaborn style
sns.set_theme(style="whitegrid")

# 1. Gender Distribution
plt.figure(figsize=(10, 6))
gender_counts = df['gender'].value_counts().head(5)
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='pastel')
plt.title('Top 5 Gender Categories of Authors (Including Co-authors)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, 'gender_distribution.png'))
plt.close()

# 2. Top 10 Nationalities
plt.figure(figsize=(10, 6))
nat_counts = df['nationality'].dropna().str.split('; ').explode().value_counts().head(10)
sns.barplot(y=nat_counts.index, x=nat_counts.values, palette='viridis')
plt.title('Top 10 Author Nationalities')
plt.xlabel('Count')
plt.ylabel('Nationality')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, 'top_nationalities.png'))
plt.close()

# 3. Top 10 Bestseller Countries
plt.figure(figsize=(10, 6))
country_counts = df['country'].value_counts().head(10)
sns.barplot(y=country_counts.index, x=country_counts.values, palette='magma')
plt.title('Top 10 Countries with Most Bestseller Entries')
plt.xlabel('Count')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, 'top_countries.png'))
plt.close()

# 4. Top 10 Publishers
plt.figure(figsize=(10, 6))
pub_counts = df['publisher'].value_counts().head(10)
sns.barplot(y=pub_counts.index, x=pub_counts.values, palette='cubehelix')
plt.title('Top 10 Publishers')
plt.xlabel('Count')
plt.ylabel('Publisher')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, 'top_publishers.png'))
plt.close()

print("Plots saved to artifacts directory.")
```
